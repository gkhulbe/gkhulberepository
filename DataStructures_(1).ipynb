## Summary of Python Data Structures and Concepts:

This notebook demonstrates several fundamental Python data structures and concepts:

### 1. Sets (`set`)
*   **Definition**: Unordered collections of unique elements.
*   **Creation**: `{item1, item2, ...}`
*   **Uniqueness**: Duplicate values are automatically removed.
*   **Unordered**: Elements do not maintain a specific order.
*   **Mutable**: Sets themselves can be modified (add/remove elements).
*   **Elements**: Can contain immutable types (numbers, strings, tuples), but not mutable types (lists, dictionaries, other sets) as direct members.
*   **Operations**: Demonstrated `add()`, `discard()`, `intersection()`, `symmetric_difference()`, and `difference()`.

### 2. Lists (`list`)
*   **Definition**: Ordered, mutable collections of items.
*   **Creation**: `[item1, item2, ...]`
*   **Mutable**: Elements can be changed, added, or removed after creation (`ls.append()`, `ls[index] = value`).
*   **Conversion**: Can be converted to/from other collections (e.g., `list(set)` or `list(tuple)`).

### 3. Tuples (`tuple`)
*   **Definition**: Ordered, **immutable** collections of items.
*   **Creation**: `(item1, item2, ...)`
*   **Immutability**: Once created, elements cannot be changed, added, or removed. Attempts to modify them directly will result in an error (as implied by the deleted cell `tp[0] = 100`).
*   **Conversion**: Can be converted to/from lists for modification purposes (`list(tuple)` then `tuple(list)`).

### 4. Mutability vs. Immutability
*   **Mutable types**: Objects whose state can be changed after creation (e.g., `list`, `set`, `dict`).
*   **Immutable types**: Objects whose state cannot be changed after creation (e.g., `int`, `float`, `str`, `tuple`). When an immutable object seems to change (like a string `st.upper()`), a *new* object is actually created.

### 5. Dictionaries (`dict`)
*   **Definition**: Unordered collections of key-value pairs.
*   **Creation**: `{key1: value1, key2: value2, ...}`
*   **Keys**: Must be unique and immutable (e.g., numbers, strings, tuples). If duplicate keys are provided during creation or update, the last assigned value for that key is retained.
*   **Values**: Can be any data type and can be duplicated.
*   **Mutable**: Dictionaries can be modified by adding, updating, or removing key-value pairs.
*   **Operations**: Demonstrated updating key-value pairs (`dt.update()`) and accessing values by key (`dt['key']`). The `fromkeys()` method was also shown for creating dictionaries with default values.
