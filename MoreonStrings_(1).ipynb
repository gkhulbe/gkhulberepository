## String Slicing with Indices

String slicing allows you to extract a portion (a substring) of a string by specifying a start and an end index. Python uses zero-based indexing, meaning the first character is at index 0, the second at index 1, and so on.

### Basic Syntax:
`string[start:end]`

*   `start`: The index where the slice begins (inclusive). If omitted, it defaults to the beginning of the string (index 0).
*   `end`: The index where the slice ends (exclusive). The character at this index is **not** included. If omitted, it defaults to the end of the string.

### Negative Indices:
You can also use negative indices, which count from the end of the string.
*   `-1` refers to the last character.
*   `-2` refers to the second to last character, and so on.

### Step Value:
You can optionally include a third parameter for the `step`:
`string[start:end:step]`

*   `step`: Determines the increment between indices. A positive step moves forward, and a negative step moves backward (useful for reversing strings).

```python
# Let's use the string 'python' for our examples
my_string = 'python'
print(f"Original string: {my_string}")

# Example 1: Basic slicing (start:end)
# Get characters from index 1 (y) up to (but not including) index 4 (o)
slice1 = my_string[1:4]
print(f"my_string[1:4]  : {slice1} (Output: 'yth')")

# Example 2: Slicing from the beginning (omitting start index)
# Get characters from the beginning up to (but not including) index 3 (h)
slice2 = my_string[:3]
print(f"my_string[:3]   : {slice2} (Output: 'pyt')")

# Example 3: Slicing to the end (omitting end index)
# Get characters from index 2 (t) to the end of the string
slice3 = my_string[2:]
print(f"my_string[2:]   : {slice3} (Output: 'thon')")

# Example 4: Using negative indices
# Get characters from the second to last (-2) up to (but not including) the last character (-1)
slice4 = my_string[-2:-1]
print(f"my_string[-2:-1]: {slice4} (Output: 'o')")

# Example 5: Using negative indices to get the last few characters
# Get characters from index -3 (h) to the end
slice5 = my_string[-3:]
print(f"my_string[-3:]  : {slice5} (Output: 'hon')")

# Example 6: Reversing a string using a negative step
# start and end are omitted, and step is -1, meaning go backwards through the entire string
reversed_string = my_string[::-1]
print(f"my_string[::-1] : {reversed_string} (Output: 'nohtyp')")

# Example 7: Slicing with a step
# Get every second character starting from the beginning
slice_step = my_string[::2]
print(f"my_string[::2]  : {slice_step} (Output: 'pto')")
```
