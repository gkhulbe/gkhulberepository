```python
# List of customers
customers = ["Alice", "Bob", "Charlie", "David", "Emma", "Frank"]

# Order details stored as tuples
orders = [
    ("Alice", "Laptop", 750, "Electronics"),
    ("Alice", "T-Shirt", 25, "Clothing"),
    ("Bob", "Vacuum Cleaner", 120, "Home Essentials"),
    ("Charlie", "Headphones", 80, "Electronics"),
    ("Charlie", "Jeans", 40, "Clothing"),
    ("David", "Blender", 60, "Home Essentials"),
    ("Emma", "Smartphone", 600, "Electronics"),
    ("Frank", "Shoes", 45, "Clothing"),
    ("Frank", "Microwave", 150, "Home Essentials")
]

# Dictionary mapping customers to ordered products
customer_orders = {
    "Alice": ["Laptop", "T-Shirt"],
    "Bob": ["Vacuum Cleaner"],
    "Charlie": ["Headphones", "Jeans"],
    "David": ["Blender"],
    "Emma": ["Smartphone"],
    "Frank": ["Shoes", "Microwave"]
}
```

```python
# 2. Classify Products by Category

# Product-category mapping
product_categories = {
    "Laptop": "Electronics",
    "T-Shirt": "Clothing",
    "Vacuum Cleaner": "Home Essentials",
    "Headphones": "Electronics",
    "Jeans": "Clothing",
    "Blender": "Home Essentials",
    "Smartphone": "Electronics",
    "Shoes": "Clothing",
    "Microwave": "Home Essentials"
}

# Unique categories using a set
unique_categories = set(product_categories.values())

print("Available Product Categories:")
for category in unique_categories:
    print(category)
```

```python
# 3. Analyze Customer Orders

# Dictionary to store customer spending
customer_spending = {}

# Calculate total spending
for order in orders:
    customer, product, price, category = order

    if customer not in customer_spending:
        customer_spending[customer] = 0

    customer_spending[customer] += price

# Classify customers
customer_classification = {}

for customer, total in customer_spending.items():
    if total > 100:
        customer_classification[customer] = "High-Value Buyer"
    elif 50 <= total <= 100:
        customer_classification[customer] = "Moderate Buyer"
    else:
        customer_classification[customer] = "Low-Value Buyer"
```

```python
# 4. Generate business insights

# Revenue per category
category_revenue = {}

for order in orders:
    customer, product, price, category = order

    if category not in category_revenue:
        category_revenue[category] = 0

    category_revenue[category] += price

# Unique products
unique_products = set()

for order in orders:
    unique_products.add(order[1])

# Customers who purchased electronics
customers_electronics = [
    customer for customer, product, price, category in orders
    if category == "Electronics"
]

# Top 3 highest-spending customers
sorted_customers = sorted(
    customer_spending.items(),
    key=lambda x: x[1],
    reverse=True
)

top_three_customers = sorted_customers[:3]
```

```python
# 5. Organize and display data

print("\nCustomer Spending Summary")

for customer, total in customer_spending.items():
    print(
        f"{customer}: ${total} - "
        f"{customer_classification[customer]}"
    )

# Customers purchasing from multiple categories
customer_categories = {}

for customer, product, price, category in orders:
    if customer not in customer_categories:
        customer_categories[customer] = set()

    customer_categories[customer].add(category)

multiple_category_customers = {
    customer for customer, categories in customer_categories.items()
    if len(categories) > 1
}

print("\nCustomers purchasing from multiple categories:")
print(multiple_category_customers)

# Customers who bought both Electronics and Clothing
customers_electronics_set = set()
customers_clothing_set = set()

for customer, product, price, category in orders:
    if category == "Electronics":
        customers_electronics_set.add(customer)

    if category == "Clothing":
        customers_clothing_set.add(customer)

common_customers = (
    customers_electronics_set.intersection(customers_clothing_set)
)

print("\nCustomers who bought both Electronics and Clothing:")
print(common_customers)
```
